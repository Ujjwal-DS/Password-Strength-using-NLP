create a custom function to split input into characters of list:

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

# TF-IDF vectorizer:

import TF-IDF vectorizer to convert styring data to numerical data:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

Apply TF-IDF vectorizer on data:

```python
X=vectorizer.fit_transform(x)
```

# Train-Test Split:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression
```

Here we will move ahead with multinomial Logistic Regression as there are more than 2 classes present

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

Lets Perform prediction over specific custom data

```python
dt=np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
clf.predict(pred)
```

Performing prediciton over test data:

```python
y_pred=clf.predict(X_test)
y_pred
```

Checking accuracy of the modcel using confusion as well as accuracy score:

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

Here we are getting approx 82% accuracy, which is not that bad.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

Precision,Recall for the above classes seems to be fair enough.

# Building Multiple Models:

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
models=[]

models.append(('LogisticRegression',LogisticRegression()))
models.append(('Random Forest',RandomForestClassifier()))
models.append(('Decision_tree',DecisionTreeClassifier()))
models.append(('KNN',KNeighborsClassifier()))
```

```python
for name,model in models:
    print(name)
    model.fit(X_train,y_train)
    
    predictions=model.predict(X_test)
    
    cm=confusion_matrix(predictions,y_test)
    print(cm)
    
    acc=accuracy_score(predictions,y_test)
    print(acc)
    print('\n')
    
```

# Thus we can say that Random Forest model to be performing the best among all the models by giving an accuracy of 98.45%.
